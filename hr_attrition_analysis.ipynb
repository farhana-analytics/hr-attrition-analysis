# Strategic Analysis of Employee Attrition & Retention Drivers


##  Objective

The objective of this analysis is to explore employee attrition patterns and identify the key factors that contribute to workforce turnover within an organisation.

By analysing employee demographic characteristics, job-related attributes, compensation levels, and workplace conditions, this study aims to understand the drivers influencing employee exit behaviour.

The insights generated from this analysis can help organisations better understand attrition trends and support data-driven strategies to improve employee retention and workforce stability.

## Dataset Overview

The dataset used in this analysis contains employee-level HR data where each row represents an individual employee within an organisation.

It includes a range of variables related to employee demographics, job roles, compensation, work experience, and workplace conditions.

Key categories of information in the dataset include:

• Demographic information (Age, Gender, Marital Status, Education)  
• Job-related attributes (Department, Job Role, Years at Company, Job Level)  
• Compensation details (Monthly Income, Salary-related indicators)  
• Work environment factors (Overtime, Job Satisfaction, Work-Life Balance)

The main variable of interest in this study is **Attrition**, which indicates whether an employee has left the organisation or continues to work within the company.

This dataset enables the exploration of relationships between employee characteristics and attrition behaviour.

## Overall Attrition Rate

## Attrition by Age

```python
sns.boxplot(x="Attrition", y="Age", data=data)
plt.title("Attrition vs Age")
plt.show()
```

### Insight

The analysis shows that employees who left the organisation tend to be slightly younger than those who stayed. This suggests that early-career employees may have a higher attrition risk compared to more experienced employees.

## Attrition by Marital Status

```python
data["MaritalStatus"].value_counts()
```

```python
pd.crosstab(data["MaritalStatus"], data["Attrition"], normalize="index") * 100
```

```python
sns.countplot(x="MaritalStatus", hue="Attrition", data=data)
plt.title("Attrition by Marital Status")
plt.show()
```

### Insight

Single employees show a higher attrition rate compared to married and divorced employees. This may indicate that employees without family commitments are more likely to explore new job opportunities.

## Attrition Risk Analysis

### Overall Attrition Rate
The overall attrition rate is **about 16.1%**, meaning about **1 in every 6 employees** left the organization.

### Key Factors Influencing Attrition

- **Marital Status:** Single employees show higher attrition (about 25%) compared to married (about 12%) and divorced employees (about 10%).
- **Gender:** Male employees show slightly higher attrition (about 17%) compared to female employees (about 15%).
- **Age:** Employees below 30 years show higher attrition (about 23 – 25%) compared to older employees (about 8–12%).
- **Monthly Income:** Employees with lower income levels show higher attrition (about 20–25%).
- **Job Role:** Roles such as **Sales Representatives and Laboratory Technicians** show higher attrition (about 20–40%).
- **Years at Company:** Employees with **0–2 years of tenure** show higher attrition (about 30%).
- **Overtime:** Employees working overtime show significantly higher attrition (about 30%) compared to employees who do not work overtime (about 10%).
- **Work-Life Balance:** Employees with poor work-life balance (Level 1) show the highest attrition (about 31%).
- **Job Satisfaction:** Employees with low job satisfaction (Level 1) show higher attrition (about 22–23%).

### Key Insight
Employees with **low job satisfaction, poor work-life balance, overtime workload, and early-career tenure** are more likely to leave the organization.

### Recommendations
- Improve **work-life balance** through flexible work schedules.
- Reduce **excessive overtime** with better workforce planning.
- Increase **job satisfaction** through employee engagement and recognition programs.
- Offer **competitive compensation** to improve employee retention.
- Provide **mentorship and career development** programs for early-career employees.

### Conclusion
Improving **work-life balance, job satisfaction, compensation, and workload management** can significantly reduce employee attrition and improve long-term employee retention.
